```python
from pdfminer.layout import LAParams, LTTextBox
from pdfminer.pdfpage import PDFPage
from pdfminer.pdfinterp import PDFResourceManager
from pdfminer.pdfinterp import PDFPageInterpreter
from pdfminer.converter import PDFPageAggregator

import re
import regex

dummy_path = '/content/drive/My Drive/DMIS/bioNLP_study/attention_is_all_you_need.pdf'
path = os.path.join(files_path, f'{input_file_name}.pdf')

pattern = r'-?(?:[^-\s]+-){2,}[^-\s]+-?'
reg = re.compile(pattern)

def fuzzy_regex(text: str, pattern: str):
    fuzzy_pattern = f'({pattern}){{e<=2}}'
    match = regex.findall(fuzzy_pattern, text, regex.BESTMATCH)
    return match 


fp = open(path, 'rb')
rsrcmgr = PDFResourceManager()
laparams = LAParams()#line_margin=1.5)#, boxes_flow=1.0)
device = PDFPageAggregator(rsrcmgr, laparams=laparams)
interpreter = PDFPageInterpreter(rsrcmgr, device)
pages = PDFPage.get_pages(fp)


grouped_text = [] 
merged_bbox_coords = []  
ori_bbox_coords = []
page_num = 0

for page in pages:
    
    print('Processing next page...')
    interpreter.process_page(page)
    page_layout = device.get_result()

    # for cv2
    page_sized_box = tuple([round(i) for i in page_layout.bbox])
    print(f'page_layout.bbox = {page_layout.bbox}')

    x1_temp, x2_temp, y1_temp, y2_temp = 0, 0, 0, 0
    text_box_idx = 0
    ori_bbox_idx = 0

    grouped_text.append([]) # add at page dimension
    merged_bbox_coords.append([])
    ori_bbox_coords.append([])

    for textbox in page_layout: # each textbox in a page 

        
    
        if isinstance(textbox, LTTextBox):
          x1_curr, y1_curr, x2_curr, y2_curr, text = textbox.bbox[0], textbox.bbox[1], textbox.bbox[2], textbox.bbox[3], textbox.get_text()
          print('At %r is text: %s' % ((x1_curr, y1_curr, x2_curr, y2_curr), text))
          # reg_res = reg.findall(text)

          ori_bbox_coords[page_num].append(list(textbox.bbox)) # tuple? list?

          if x1_temp != 0 and (abs(x1_curr-x1_temp) < 50) and (abs(y2_temp-y1_curr) < 100): # added condition regarding y value
            text_ori = grouped_text[page_num][text_box_idx]
            grouped_text[page_num][text_box_idx] = text_ori + text 
            print(f'grouped_text[page_num][text_box_idx]={grouped_text[page_num][text_box_idx]}')

            # merged_bbox_coords[page_num][text_box_idx][2:] = x2_curr, y2_curr # bounding box 우하단 좌표 update. 
            

            if x2_curr > x2_temp+10:
              merged_bbox_coords[page_num][text_box_idx][2:] = x2_curr, y2_curr
            else:
              merged_bbox_coords[page_num][text_box_idx][3] = y2_curr        

          else: # should be treated separately from the prev textbox
            if x1_temp != 0: 
              text_box_idx +=1
            x1_temp, x2_temp, y1_temp, y2_temp = x1_curr, x2_curr, y1_curr, y2_curr
            grouped_text[page_num].append(text) # add at bbox dimension
            new_bbox_coords = [x1_curr, y1_curr, x2_curr, y2_curr] # 3차원 됨 리스트를 쓰면. 페이지 -> bbox -> 좌표
            merged_bbox_coords[page_num].append(new_bbox_coords)

            ori_bbox_idx +=1

    page_num +=1

##############################################################################
# add regex for finding example number and iupac identifier pair 

# find table -> check if it contains ic50 values
# 



res = []

for page_idx in range(len(grouped_text)): 
  res.append([])
  for bbox_idx in range(len(grouped_text[page_idx])):

    bounded_text_stripped = re.sub('(\s)*(\n)+(\s)*', '', ''.join(grouped_text[page_idx][bbox_idx])) # join text in grouped_text and get rid of whitespaces -> 아 여기서 내가 trailing dash를 없애버린건가?
    print(f'repr(bounded_text_stripped) = {repr(bounded_text_stripped)}') 

    reg = re.compile(pattern)
    re_res = reg.findall(bounded_text_stripped)
    res[page_idx].append(re_res)


    # regex_res = fuzzy_regex(bounded_text_stripped, pattern) 
    # res[page_idx][bbox_idx] = regex_res.group() # stores match result in list. or you could do a dictionary instead.


```
